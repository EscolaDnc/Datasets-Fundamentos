```python
def binarysearch(arr, alvo):
  low = 0
  high = len(arr) - 1
  while low <= high:
    mid = (low + high) // 2
    tentativa = arr[mid]
    if tentativa == alvo:
      return mid
    if tentativa > alvo:
      high = mid - 1
    else:
      low = mid + 1
  return None
```

```python
lista = [2, 4, 8, 17, 25, 27, 30, 10, 32, 38, 45]
alvo = 38
print('lista: ', lista)
print('alvo: ', alvo)
print('resultado: ', binarysearch(lista, alvo))
```
